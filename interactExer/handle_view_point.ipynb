```julia
using Plots
using Interact
```

- See documentation of Plots.jl to see whole attribute : https://docs.juliaplots.org/latest/attributes/
- See [Setting the Viewpoint with Azimuth and Elevation](https://jp.mathworks.com/help/matlab/creating_plots/setting-the-viewpoint-with-azimuth-and-elevation.html) if you are first time to see words `azimuthal` and `elevation`
- See documentation of Interact.jl : https://juliagizmos.github.io/Interact.jl/latest/

```julia
xs, ys, zs = rand(10), rand(10), rand(10)

colors=OrderedDict(
    "red" => :red, 
    "blue" => :blue,
    "green" => :green
)

markershapes=[
    :none,
    :auto,
    :circle,
    :rect,
    :star5,
    :diamond,
    :hexagon,
    :cross,
    :xcross,
    :utriangle,
    :dtriangle,
    :rtriangle,
    :ltriangle,
    :pentagon,
    :heptagon,
    :octagon,
    :star4,
    :star6,
    :star7,
    :star8,
    :vline,
    :hline,
    :+,
    :x
]

xs,ys,zs=rand(10),rand(10),rand(10)

app = @manipulate for azimuthal in 0:90,
                      elevation in 0:90,
                      color in togglebuttons(colors),
                      mshape in dropdown(markershapes)
     scatter3d(xs, ys, zs,
        cam = (azimuthal, elevation),
        title = "cam=($azimuthal,$elevation)",
        xlabel = "x",
        ylabel = "y",
        zlabel = "z",
        color=color,
        markershape=mshape,
    )
end

ui = @layout! app vbox(hbox(:azimuthal,:elevation),hbox(:color,hskip(9em),:mshape),vskip(4em),observe(_))
```

```julia
using Mux
WebIO.webio_serve(page("/", req -> ui),9999)
```
